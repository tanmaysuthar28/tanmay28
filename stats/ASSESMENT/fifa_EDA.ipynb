```python
top_10_players_by_wage = df.sort_values('Wage', ascending=False)[['Name', 'Wage']][:10].set_index('Name')

fig, ax = plt.subplots(figsize=(12, 8))
bars = top_10_players_by_wage.plot(kind='bar', ax=ax, color=sns.color_palette("cool", 1))

ax.set_xlabel('Player', fontsize=14, fontweight='bold')
ax.set_ylabel('Wage', fontsize=14, fontweight='bold')
ax.set_title('Top 10 Players by Wage', fontsize=16, fontweight='bold')

plt.xticks(rotation=45, fontsize=12)

ax.grid(True, linestyle='--', alpha=0.6)

for bar in bars.containers[0]:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2 - 0.1, yval + 0.5, int(yval), fontsize=12, fontweight='bold', color='black')

plt.tight_layout()
plt.show()
```

```python
positions = ['RF','LF','CF','CDM','CAM','CB','GK']
positions = sorted(positions)

by_position = df.set_index(['Position', 'Name'], drop=True).sort_index(level=0)

position_g = by_position[['Value']].loc[positions].groupby(level=0)


position_dfs = []
for n,g in position_g:
    position_dfs.append(g.sort_values(by='Value', ascending=False).iloc[:5])
```

```python
color= sns.color_palette("Set3", len(position_dfs))
fig, ax = plt.subplots(len(position_dfs), 1, figsize=(10,20))
for i, (df_1, pos) in enumerate(zip(position_dfs,positions)):
    df_1.droplevel(0).plot.bar(ax=ax[i],title=pos, xlabel="", color=color[i], ylabel='Euros')
    ax[i].tick_params(axis='x',rotation=0)

plt.tight_layout()
```
